```
import ezodf


def abrir(arquivo):
    documento = ezodf.opendoc(arquivo)
    return documento

def ler(documento):
    planilhas = documento.sheets
    return planilhas

def modificarIdentificacao(planilhas, lotes, n_lote):
    denominacao = lotes[n_lote].name
    municipio = lotes[n_lote]['B1'].value
    planilhas[0]['B10'].set_value(denominacao)
    planilhas[0]['B17'].set_value(municipio)

def lerPontos(lotes, n_lote):
    n_linhas = lotes[n_lote].nrows()
    n_pontos = 0
    listaP = []
    x = 2
    while x <= n_linhas:
        ponto = lotes[n_lote]['A{}'.format(x)].value
        if ponto != None:
            listaP.append(lotes[n_lote]['A{}'.format(x)].value)
        else:
            n_pontos += 1
        x += 1
    return listaP

def buscarPonto(pontos, pontoBuscado):
    p_total = pontos[0].nrows()
    x = 1
    while x <= p_total:
        observado = pontos[0]['A{}'.format(x)].value
        if pontoBuscado == observado:
            break

        x += 1
    return x

def buscarPontos(pontos, listaPontos):
    indexPontos = []
    for x in listaPontos:
        indexPontos.append(buscarPonto(pontos, x))

    return indexPontos

def inserirPonto(pontos, index, planilha, linhaInsercao, lotes, indexInserido):
    codigo = pontos[0]['A{}'.format(index)].value
    long =  pontos[0]['B{}'.format(index)].value
    sigmaLong = pontos[0]['C{}'.format(index)].value
    lat = pontos[0]['D{}'.format(index)].value
    sigmaLat = pontos[0]['E{}'.format(index)].value
    alt = pontos[0]['F{}'.format(index)].value
    sigmaAlt = pontos[0]['G{}'.format(index)].value
    metodoPosicionamento = pontos[0]['H{}'.format(index)].value
    tipoLimite = pontos[0]['I{}'.format(index)].value
    #cns = pontos[0]['J{}'.format(index)].value
    #matricula = pontos[0]['K{}'.format(index)].value
    descritivo = lotes[0]['B{}'.format(indexInserido)].value

    planilha[1]["A{}".format(linhaInsercao)].set_value(codigo)
    planilha[1]["B{}".format(linhaInsercao)].set_value(long)
    planilha[1]["C{}".format(linhaInsercao)].set_value(sigmaLong)
    planilha[1]["D{}".format(linhaInsercao)].set_value(lat)
    planilha[1]["E{}".format(linhaInsercao)].set_value(sigmaLat)
    planilha[1]["F{}".format(linhaInsercao)].set_value(alt)
    planilha[1]["G{}".format(linhaInsercao)].set_value(sigmaAlt)
    planilha[1]["H{}".format(linhaInsercao)].set_value(metodoPosicionamento)
    planilha[1]["I{}".format(linhaInsercao)].set_value(tipoLimite)
    #planilha[1]["J{}".format(linhaInsercao)].set_value(cns)
    #planilha[1]["K{}".format(linhaInsercao)].set_value(matricula)
    planilha[1]["L{}".format(linhaInsercao)].set_value(descritivo)

def inserirPontos(pontos, planilha, lotes, indexPontos ):
    w = 2
    inicio = 12
    for y in indexPontos:
        inserirPonto(pontos, y, planilha, inicio, lotes, w)
        w += 1
        inicio += 1

#Abre os arquivos
modelo = abrir("modelo.ods")
assentamento = abrir('assentamento.ods')
processamento = abrir('processamento.ods')

#Ler as planilhas
planilhas = ler(modelo)
lotes = ler(assentamento)
pontos = ler(processamento)

#Modifica Identificacao
modificarIdentificacao(planilhas, lotes, 0)

#Ler os Pontos do Lote
pontosLote = lerPontos(lotes, 0)

#Buscar indices dos Pontos do lote na planilha de processamento
indexPontos = buscarPontos(pontos, pontosLote)

#Inserir Todos os Pontos na planilha modelo
inserirPontos(pontos, planilhas, lotes, indexPontos)

#Salva os dadps em um novo arquivo
modelo.saveas("lote_01.ods")
```
